```python
import pandas as pd
import numpy as np
```

```python
df_1 = pd.read_csv('/content/drive/MyDrive/amazon_reviews.csv')
df_2 = pd.read_csv('/content/drive/MyDrive/amazon_reviews_2.csv')
df_3 = pd.read_csv('/content/drive/MyDrive/amazon_reviews_3.csv')
```

```python
df = pd.concat([df_1,df_2,df_3],ignore_index=True)
```

```python
df.head()
```

```python
df.shape,df_1.shape,df_2.shape,df_3.shape
```

```python
## Renaming columns for consistency
df.rename(columns={'reviews.username':'user_id','asins' : 'product_id','reviews.rating':'rating'},inplace=True)
```

```python
df = df[['user_id','product_id','rating']]
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True) # Dropping null values
```

```python
df.drop_duplicates(inplace=True) # Dropping duplicate values
```

```python
df['product_id'] = df['product_id'].astype('str')
df['product_id'] = df['product_id'].str.split(',')
```

```python
df = df.explode('product_id') # One row per product
```

```python
user_count = df['user_id'].value_counts()
product_count = df['product_id'].value_counts()
```

```python
#Filtering out users/products with < 5 interactions
valid_users = []
```

```python
for users,count in user_count.items():
  if count >5:
    valid_users.append(users)
```

```python
valid_products =[]
```

```python
for users,count in product_count.items():
  if count >5:
    valid_products.append(users)
```

```python
df = df[df['user_id'].isin (valid_users)]
df = df[df['product_id'].isin (valid_products)]
```

```python
df.reset_index(drop=True, inplace=True)
```
